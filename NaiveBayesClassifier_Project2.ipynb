## CHURN ANALYSIS USING  NAIVE BAYES

```python
# Import Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
#Load Dataset
df=pd.read_excel("churn_dataset.xlsx")
df.head()
```

```python
#Preprocessing
df['Churn'] = df['Churn'].str.strip().str.capitalize()
df['Sex'] = df['Sex'].map({'Male': 1, 'Female': 0})
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
#splitting data 

X = df[['Age', 'Tenure', 'Sex']]
y = df['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(df['Churn'].unique())
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Churn'])
```

#  Gaussian Naive Bayes 


```python
#training Model
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import joblib
joblib.dump(model, 'naive_bayes_churn_model.pkl')
```

```python
# Test model by input data
def predict_churn(age, tenure, sex):  # sex: 1 for Male, 0 for Female
    input_data = pd.DataFrame([[age, tenure, sex]], columns=['Age', 'Tenure', 'Sex'])
    prediction = model.predict(input_data)[0]
    return prediction

# اختبار سريع
print("Test Prediction (35, 12, Male):", predict_churn(35, 12, 1))
```
